# Count Rotations

```python
def evaluate_test_cases(
count_rotations, tests):
    case_number = 0
    passed = 0
    failed = 0
    for test in tests:
        print("#CASE", 
              case_number, "\n\n")
        print("Input:\n",
              "nums:",
              test['input']['nums'],
             "\n\n")
        print("Expected output: ",
              test['output'], "\n\n")
        
        #passing in the cases.
        nums = test['input']['nums']
        output = test['output']
        result = count_rotations(nums) 
        #^^ ACTUAL OUTPUT ^^
        
        print("Actual output: ",
              result, "\n\n")
        #condition for case validation.
        if result == output:
            print("result:PASSED", "\n\n\n")
            passed += 1
        else:
            print("result:FAILED", "\n\n\n")
            failed += 1
            
         
        if case_number == (len(tests)-1):
            print("#SUMMARY\n")
            print("TEST CASES:", case_number + 1,
                  "\n"
                  "PASSED:", passed,
                  "\n"
                  "FAILED:", failed)
        
        #increment case_number.
        case_number += 1
        
        
```

# Rotated list Searching

```python
def evaluate_test_cases(get_position
                        , tests):
    case_number = 0
    passed = 0
    failed = 0
    for test in tests:
        print("#CASE", 
              case_number, "\n\n")
        print("Input:\n",
              "nums:",
              test['input']['nums'],"\n",
              "Searched number:",
              test['input']['searched_num'],
             "\n\n")
        print("Expected output: ",
              test['output'], "\n\n")
        
        #passing in the cases.
        nums = test['input']['nums']
        searched_num = test['input']['searched_num']
        output = test['output']
        result = get_position(nums, searched_num) 
        #^^ ACTUAL OUTPUT ^^
        
        print("Actual output: ",
              result, "\n\n")
        #condition for case validation.
        if result == output:
            print("result:PASSED", "\n\n\n")
            passed += 1
        else:
            print("result:FAILED", "\n\n\n")
            failed += 1
            
         
        if case_number == (len(tests)-1):
            print("#SUMMARY\n")
            print("TEST CASES:", case_number + 1,
                  "\n"
                  "PASSED:", passed,
                  "\n"
                  "FAILED:", failed)
        
        #increment case_number.
        case_number += 1
        
        
```
