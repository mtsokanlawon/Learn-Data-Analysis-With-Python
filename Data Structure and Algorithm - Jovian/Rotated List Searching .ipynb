# 1.PROBLEM:
    
We need to find the position of a target number in a list of numbers rotated an unknown number of times, assuming all the numbers are unique.

# 2.Define input and output 

input: A list of numbers e.g
nums(rotated_list) = `[9,11,13,1,3,5,7]`

#sorted_list = `[1,3,5,7,9,11,13]`
and; number to be searched:
searched_num = `3`

output:position of the number searched. To better understand this, let's have:

nums = `[9,11,13,1,3,5,7]`

#let us assume we are searching for `3`, the position of `3` is `index position 4`.

```python
#signature of function 
def get_position(nums, searched_num):
    position = 0
    while position != len(nums):
        #check position for 
        #searched number.
        if nums[position] == searched_num:
            return position
        #increment position
        position += 1
    #searched number not found.
    return "Not found!"
```

# 4.Solution Description 


 This problem can be solved using the Linear and Binary Search.

Using linear Search:

A function that will get the position of the target number is created, the function reveives the rotated list of numbers and the target number as input.

How it works:

1.A variable position is created; it receives the current index position iterated in the list(starting from `0` rightwards).

2.We have a condition: While the list is not exhausted, the current position is checked.

-if the number matches the target number, the current index position is returned.

-if not, the variable position is invremented and the search process is repeated till the list is exhausted.

-if the target number is yet to be found, then `'Not found!'` should be returned.


```python
tests = [test,test1,test2,test3,
        test4,test5,test6,test7]
```

```python
#evaluate test function.
def evaluate_test_cases(get_position
                        , tests):
    case_number = 0
    passed = 0
    failed = 0
    for test in tests:
        print("#CASE", 
              case_number, "\n\n")
        print("Input:\n",
              "nums:",
              test['input']['nums'],"\n",
              "Searched number:",
              test['input']['searched_num'],
             "\n\n")
        print("Expected output: ",
              test['output'], "\n\n")
        
        #passing in the cases.
        nums = test['input']['nums']
        searched_num = test['input']['searched_num']
        output = test['output']
        result = get_position(nums, searched_num) 
        #^^ ACTUAL OUTPUT ^^
        
        print("Actual output: ",
              result, "\n\n")
        #condition for case validation.
        if result == output:
            print("result:PASSED", "\n\n\n")
            passed += 1
        else:
            print("result:FAILED", "\n\n\n")
            failed += 1
            
         
        if case_number == (len(tests)-1):
            print("#SUMMARY\n")
            print("TEST CASES:", case_number + 1,
                  "\n"
                  "PASSED:", passed,
                  "\n"
                  "FAILED:", failed)
        
        #increment case_number.
        case_number += 1
        
                      
```

Now we can test the cases at once.

```python
evaluate_test_cases(get_position,
                        tests)
```

# 6.4 Solution Description 

```python
#Our function for binary search.
def get_position_binary(nums, searched_num):
    lo = 0
    hi = len(nums)-1
    
   
    while hi == 0 or hi > 0:
        mid = (lo + hi)//2
        mid_number = nums[mid]
        
        print("lo:",lo, "hi:",hi,
              "mid:",mid, "mid_number:",
              mid_number)
        
        #Note that the order of the 
        #following conditional 
        #statements are important 
        #in getting adequate result.
        if mid_number == searched_num:
            return mid
        
        #check if answer lies to
        #the right.
        elif mid_number < searched_num and nums[len(nums)-1] > (searched_num-1):
            lo = mid + 1
        #the left.
        elif mid_number > searched_num and nums[0] < (searched_num+1):
            hi = mid - 1
        #the right.
        elif mid_number > searched_num and nums[len(nums)-1] > (searched_num-1):
            lo = mid + 1
        #the left
        else:
            hi = mid - 1
            
    return "Not found!"
```

```python
#evaluate test function.
def evaluate_test_cases(get_position_binary
                        , tests):
    case_number = 0
    passed = 0
    failed = 0
    for test in tests:
        print("#CASE", 
              case_number, "\n\n")
        print("Input:\n",
              "nums:",
              test['input']['nums'],"\n",
              "Searched number:",
              test['input']['searched_num'],
             "\n\n")
        print("Expected output: ",
              test['output'], "\n\n")
        
        #passing in the cases.
        nums = test['input']['nums']
        searched_num = test['input']['searched_num']
        output = test['output']
        result = get_position_binary(nums, searched_num) 
        #^^ ACTUAL OUTPUT ^^
        
        print("Actual output: ",
              result, "\n\n")
        #condition for case validation.
        if result == output:
            print("result:PASSED", "\n\n\n")
            passed += 1
        else:
            print("result:FAILED", "\n\n\n")
            failed += 1
            
         
        if case_number == (len(tests)-1):
            print("#SUMMARY\n")
            print("TEST CASES:", case_number + 1,
                  "\n"
                  "PASSED:", passed,
                  "\n"
                  "FAILED:", failed)
        
        #increment case_number.
        case_number += 1
        
                      
```
